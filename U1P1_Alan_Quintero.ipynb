Recorrido Simple (explicacion liena a liena)
- definir una funcion llamada "recorrido_simple", recibe un parametro llamado lista
- inicializa una variable llamada "total" asignandole un valor de 0
- inicia una iteracion con for, luego cun una variable temporal de cada elemento(x), in lista indica que la iteracion se hara en los elementos contenidos en la variable lista
- abreviacion de la sentencia: total + x = total
- regresa el valor final de total

```python
def recorrido_simple(lista):
    total = 0
    for x in lista:
        total += x
    return total
```

Doble Ciclo (explicacion linea a linea)
- definir una funcion llamada "doble_ciclo", recibe un parametro llamado lista
- inicializa una varable llamada "contador" asignandole un valor de 0
- inicia una iteracion con for, luego con una variable temporal de cada elemento(i), in range(len(lista)) calcula la longitud de lista paa luego usar esa longitud para generar indices numericos de 0 en adelante   
- inicia una iteracion con for, luego con una variable temporal de cada elemento(j), in range(len(lista)) calcula la longitud de lista paa luego usar esa longitud para generar indices numericos de 0 en adelante   
- abreviacion de la sentencia: contador + lista[i] * lista[j] = contador, se multiplican los indices en i y j, se suman a el valor de contador
- regresa el valor final de contador

```python
def doble_ciclo(lista):
    contador = 0
    for i in range(len(lista)):
        for j in range(len(lista)):
            contador += lista[i] * lista[j]
    return contador
```

Experimento (explicacion liena a linea)
- se define una variable "experimento"
- se inicializa una variavle "tamanios" con un arreglo de 4 valores: 100, 500, 1000 y 2000
- se imprime un mensaje haciendo alucion a una tabla con los apartados: tamanios, recorrido simple y doble ciclo
- se imprime un mensaje para separar(ayuda a la estetica)
- se inicia una iteracion con for, con una varable temporal para cada elemento(n) dentro de la variable tamanios
- dentro del for, se inicia otra variable "datos" con el valor creado de una lista que contenga "n" numeros enteros pseudoaleatorios, donde cada numero puede ir desde el 1 al 100 incluyendolos
- se declara la variable "inicio" cuyo valor es una variale que mida el tiempo de ejecucion
- a "recorrido_simple" le pasamos como parametro la lista datos
- con una variable "t1" con time.time() devuelve el tiempo actual, se le resta inicio donde se guardo el resultado a la primera llamada de time.time(), casi de forma recursiva
- volvemos a tomar el tiempo con "inicio"
- ahora le pasamos la lista datos a doble_ciclo
-  con una variable "t2" con time.time() devuelve el tiempo actual, se le resta inicio donde se guardo el resultado a la primera llamada de time.time(), casi de forma recursiva
- se imprime con f-string, permitiendo incrustar expresiones, n es la variable cuyo valor se imprimirá, :6d indica el formato de impresion de un numero entero, con 6 decimales.
- se separa literalmente usando |
- con los siguientes: t1 y t2, son las variables cuyos valores se imprimiran, para t1, se formateara como un flotante ocupando un ancho de 20 caracteres mostrando 6 decimales. Y para t2 se formateara como un flotante ocupando un ancho de 15 caracteres mostrando 6 decimales.
- finalmente se asegura que experimento() solo se ejecute cuando el archivo se inicia como el programa principal.

```python
def experimento():
    tamanios = [100, 500, 1000, 2000]
    print("Tamano | Recorrido simple (s) | Doble ciclo (s)")
    print("----------------------------------------------")

    for n in tamanios:
        datos = [random.randint(1, 100) for _ in range(n)]

        inicio = time.time()
        recorrido_simple(datos)
        t1 = time.time() - inicio

        inicio = time.time()
        doble_ciclo(datos)
        t2 = time.time() - inicio

        print(f"{n:6d} | {t1:20.6f} | {t2:15.6f}")

if __name__ == "__main__":
    experimento()
```
